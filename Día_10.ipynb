```
import matplotlib.pyplot as plt
```

```
N = 100
X = []
Y = []

for i in range(1,N+1):
  X.append(i)
  Y.append(i)
```

#Código tema 12_2

```
#Importantes paquetes
import matplotlib.pyplot as plt 
#import math as mt
from math import * #importar todo
#funciones matemáticas a graficar 
#0.8sin(200-2πx)
#1.1cos(250-πx)

X = []
Y_sin = []
Y_cos = []
x = -10

while(x <= 10):
  X.append(x)
  Y_sin.append(0.8*sin(200-2*pi*x))
  Y_cos.append(1.1*cos(250-pi*x))
  x += 0.1
```

```
print(X)
print(Y_sin)
print(Y_cos)
```

```
plt.plot(X,Y_sin)
```

```
plt.plot(X,Y_cos)
```

```
plt.plot(X,Y_sin, label = '0.8sin(200-2πx)')
plt.plot(X,Y_cos, label = '1.1cos(250-πx)')
plt.legend()
```

```
Y_sup = []
for n in range(len(X)):
  Y_sup.append(Y_sin[n]+Y_cos[n])
print(Y_sup)
```

```
plt.plot(X,Y_sup, label = 'Pulsos de ondas')
plt.legend()
```

#Código tema 12_3

```
#Importar paquetes
import numpy as np
import matplotlib.pyplot as plt
```

```
def abrir_arc(nombre_archivo):
  f = open(nombre_archivo,'r')
  return f
```

```
def datos_a_matriz(f):
  matriz = []
  matriz = [line.split() for line in f] #se va a particionar en líneas
  #de forma iterativa: se crea un arreglo
  return matriz
```

```
def cerrar_archivo(f):
  f.close()

#La siguiente función hace una conversión de variables, pues el 
#archivo tiene datos tipo str, lo cual no se puede graficar.
def str_a_float(matriz): 
  for n in range(len(matriz)):
    matriz[n][0] = float(matriz[n][0])
    matriz[n][1] = float(matriz[n][1])
```

```
#Esta función hace un arreglo en la lista, para tener vectores independientes
#que se puedan relacionar x con y
def matriz_numpy(matriz):
  matriz = np.array(matriz)
  return matriz
```

```
def graficar(matriz):
  plt.plot(matriz[:,0], matriz[:,1],linewidth = 0.5) 
  #los dos puntos colocan todos los elementos en vertical
  plt.scatter(matriz[:,0], matriz[:,1], c = 'r', s = 1) 
  #scatter coloca puntos asociados a cada coordenada,s:size,c:color
```

```
f = abrir_arc('Archivo_dato.txt')
matriz = datos_a_matriz(f)
cerrar_archivo(f)
str_a_float(matriz)
matriz = matriz_numpy(matriz)
graficar(matriz)
```
